```python
cur_url=browser.current_url
print(cur_url)
```

```python
headers = {"User-Agent": "Mozilla/5.0"}
request = req.Request(cur_url, headers=headers)
response = req.urlopen(request)
```

```python
parse_html = BeautifulSoup(response, 'html.parser')
```

```python
tables = parse_html.find_all('table')[0]
tr = tables.find_all('tr')
a = tables.find_all('a')
# a
```

```python
url_lists = []
for i in a:
    player_url = 'https://www.vleague.jp' + i.attrs['href']
#     print(player_url)
    url_lists.append(player_url)
print(url_lists)
```

```python
number_list = []
player_list = []
for i in tr[1:]:
    td = i.find_all('td')
    number = td[0].text
    number_list.append(number)
    player = td[1].text.replace('\xa0', ' ')
    player_list.append(player)
print(number_list, player_list)
```

```python
team_dir = '/Users/sasno/Desktop/MyPandas/vleague/{}'.format(team)

if not os.path.isdir(team_dir):
    os.makedirs(team_dir)
```

```python
for i in range(len(player_list)):
    html = requests.get(url_lists[i], headers=headers)
    data = pd.read_html(html.text, header = [0, 1])
    stats = pd.DataFrame(data[0])
    stats = stats.rename({"チ｜ム":"チーム", "サ｜ブ":"サーブ"}, axis=1)
    stats.to_csv('/Users/sasno/Desktop/MyPandas/vleague/{0}/{1}_{2}_{3:%Y}-{3:%m}-{3:%d}.csv'.format(team, number_list[i], player_list[i], now), index=False, encoding='cp932')
```
